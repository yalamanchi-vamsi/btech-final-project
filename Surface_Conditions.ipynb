```python
#surface conditions
import pandas as pd
import matplotlib.pyplot as plt
# setting figure size manually
plt.rcParams["figure.figsize"] = (22,5)
#data=pd.read_csv(r'C:\Users\exam2\Desktop\ACCIDENTS.csv',encoding='utf-8')
data=pd.read_csv(r'C:\Users\Eliyazz\Desktop\IMPORTANT\Project\FARS2007NationalCSV\ACCIDENT.csv')
df=pd.DataFrame(data)
#required columns
cols3=['SUR_COND','FATALS','PERSONS']
dfObj3=pd.DataFrame(df[cols3])
#total fatals for different groups of surface condition
dataf3=dfObj3.groupby(['SUR_COND']).sum().sum(level=['SUR_COND']).reset_index()
#counting total fatals
count3=dataf3.FATALS.sum()
#total count
print(count3)
#using lamda apply function to create a new column
dataf3['FATAL_PERCENTAGE_SURFACE']=dataf3.apply(lambda row:(row.FATALS/count3)*100,axis=1)
#refined data
print(dataf3)
#defining x-limits
plt.xlim((0,10))
#defining y-limits
plt.ylim((0,100))
s=['','Dry','Wet','Snow/Slush','Ice/Frost','Sand,Dirt,Mud,Gravel','Water(standing/moving)','Oil','Other','Unknown']
#range between each interval
plt.yticks(range(0,100,10))
plt.xticks(range(0,10,1),s)
#x-label
plt.xlabel("Surface Conditions")
#y-label
plt.ylabel("Percentage(%)")
#title
plt.title("(e)Fatal accidents on different surface conditions")
#plotted-graph
plt.bar(dataf3['SUR_COND'],dataf3['FATAL_PERCENTAGE_SURFACE'],align='center',color='VIOLET',width=0.4)
plt.show()
#
```
